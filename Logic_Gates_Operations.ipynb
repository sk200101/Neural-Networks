# Logical Operations

```python
class Logic_Operations:
    def __init__(self,x,y):
        self.x = x
        self.y = y
        
    def AND(self):
        print("     Truth Table  for AND           ")
        print("           x | y | AND")
        for i in range(len(self.x)):
            value = self.x[i] and self.y[i]
            print("           " + str(self.x[i]) + " | " + str(self.y[i])+ " |  " + str(value))
        
    def OR(self):
        print("     Truth Table  for OR           ")
        print("           x | y | OR")
        for i in range(len(self.x)):
            value = self.x[i] or self.y[i]
            print("           " + str(self.x[i]) + " | " + str(self.y[i])+ " |  " + str(value))
        
    def NOT(self):
        print("     Truth Table  for NOT           ")
        print("           x |NOT| y | NOT")
        for i in range(len(self.x)):
            value_1 = not self.x[i]
            if (value_1) == 1:
                value_1 = 1
            else:
                value_1 = 0
            value_2 = not self.y[i]
            if (value_2) == 1:
                value_2 = 1
            else:
                value_2 = 0
            print("           " + str(self.x[i]) + " | " + str(value_1) + " | " + str(self.y[i])+ " |  " + str(value_2))
        
    def XOR(self):
        print("     Truth Table  for XOR           ")
        print("           x | y | XOR")
        for i in range(len(self.x)):
            value = self.x[i] ^ self.y[i]
            print("           " + str(self.x[i]) + " | " + str(self.y[i])+ " |  " + str(value))
        
        
z = Logic_Operations([0,0,1,1],[0,1,0,1])
print("1 for AND")
print("2 for OR")
print("3 for NOT")
print("4 for XOR")
print("0 for exit")
n = -1
while(n != 0):
    n = int(input("Enter your choice: "))
    if n == 1:
        z.AND()
        print()
    elif n == 2:
        z.OR()
        print()
    elif n == 3:
        z.NOT()
        print()
    elif n == 4:
        z.XOR()
        print()

print("Thank You!!!")
```
